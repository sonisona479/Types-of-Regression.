**1) Linear Regression**

```python
import plotly.express as px
import pandas as pd
from sklearn.linear_model import LinearRegression

X = pd.Series([1, 2, 3, 4, 5])
y = pd.Series([2, 4, 6, 8, 10])

model = LinearRegression()
model.fit(X.values.reshape(-1, 1), y)
y_pred = model.predict(X.values.reshape(-1, 1))

df = pd.DataFrame({"X": X, "Actual": y, "Predicted": y_pred})
fig = px.scatter(df, x="X", y="Actual", title="Linear Regression", labels={"X": "Input", "Actual": "Actual Output"})
fig.add_scatter(x=df["X"], y=df["Predicted"], mode="lines", name="Regression Line")
fig.show()
```

**2) Polynomial Regression**

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 5, 10, 17, 26])

poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, y)
y_pred = model.predict(X_poly)

sns.set_style("darkgrid")
plt.figure(figsize=(8, 5))
sns.scatterplot(x=X.flatten(), y=y, color="blue", s=80, label="Actual Data")
sns.lineplot(x=X.flatten(), y=y_pred, color="red", label="Polynomial Fit")
plt.title("Polynomial Regression with Smooth Curve")
plt.legend()
plt.show()
```

**3) Ridge Regression**

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 6, 8, 10])

model = Ridge(alpha=1.0)
model.fit(X, y)
y_pred = model.predict(X)

plt.figure(figsize=(8, 5))
sns.regplot(x=X.flatten(), y=y, scatter_kws={"s": 80}, line_kws={"color": "red"}, ci=95)
plt.title("Ridge Regression with Confidence Interval")
plt.show()
```

**4) Lasso Regression**

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 6, 8, 10])

model = Lasso(alpha=0.1)
model.fit(X, y)
y_pred = model.predict(X)

residuals = y - y_pred

plt.figure(figsize=(8, 5))
sns.histplot(residuals, bins=10, kde=True, color="purple")
plt.title("Lasso Regression Residual Plot")
plt.xlabel("Residuals")
plt.ylabel("Frequency")
plt.show()
```

**5) ElasticNet Regression**

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import ElasticNet

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 6, 8, 10])

model = ElasticNet(alpha=0.1, l1_ratio=0.5)
model.fit(X, y)
y_pred = model.predict(X)

plt.figure(figsize=(8, 5))
sns.heatmap(np.corrcoef(y, y_pred), annot=True, cmap="coolwarm")
plt.title("ElasticNet Regression Correlation Heatmap")
plt.show()
```

**6) Support Vector Regression**

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.svm import SVR

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 6, 8, 10])

model = SVR(kernel='rbf')
model.fit(X, y)
y_pred = model.predict(X)

fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X.flatten(), y, y_pred, color="blue", label="Actual Data")
ax.plot(X.flatten(), y_pred, color="red", label="SVR Fit")
ax.set_title("Support Vector Regression 3D Surface Plot")
plt.legend()
plt.show()
```

**7) Decision Tree Regression**

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 6, 8, 10])

model = DecisionTreeRegressor()
model.fit(X, y)

plt.figure(figsize=(10, 6))
plot_tree(model, filled=True, feature_names=["X"], rounded=True)
plt.title("Decision Tree Regression Visualization")
plt.show()
```
